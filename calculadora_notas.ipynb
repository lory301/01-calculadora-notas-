# Calculadora de notas

```python
nome = input("Nome do aluno: ")

nota1 = float(input("Nota 1: "))
nota2 = float(input("Nota 2: "))

media = (nota1 + nota2) / 2

if media >= 7:
    resultado = "Aprovado"
else:
    resultado = "Reprovado"

print(nome)
print(media)
print(resultado)
```
